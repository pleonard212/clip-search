```python
from sentence_transformers import SentenceTransformer, util
from PIL import Image
import torch
```

```python
model = SentenceTransformer('clip-ViT-B-32')
```

```python
from PIL import Image
import urllib.request
import io

URL = 'https://navn.ku.dk/maanedens_navn/dansk-vestindien/Christiansted_Sabroe_1839_Kgl_Bibl.jpg'

with urllib.request.urlopen(URL) as url:
    f = io.BytesIO(url.read())

img = Image.open(f)

display(img)
```

```python
img_emb = model.encode(Image.open(f))
```

```python
strings = ['A picture of London at night', \
           'A flag in the harbor', \
           'Oscar the Grouch', \
           'A flag waves above a tropical bay, mountain in the distance', \
           'A cat on a table', \
          ]
text_emb = model.encode(strings)
```

```python
cos_scores = util.cos_sim(img_emb, text_emb)
```

```python
actualscores = cos_scores.numpy()
results = {}

for i, score in enumerate(actualscores[0]): 
    results[i] = score

# print(results)
for k,v in sorted(results.items(), key=lambda item: item[1], reverse=True):
    print(str(round((v) * 100)) + '%: ' + strings[k])
```
